```python
# Group the data by the manufacturer column.
manufacturer_counts = df['manufacturer'].value_counts()

```

```python
manufacturer_counts
```

```python
# Sorting in Dec order as we need the top bought/Highest frequency one
sorted_manufacturer_counts = manufacturer_counts.sort_values(ascending=False)

# showing the top 5
top_manufacturers = sorted_manufacturer_counts.head()
```

```python
top_manufacturers
```

1) These are the top Manufacturers in the dataset, getting the most orders

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame of the top manufacturers.
#default pe descending
top_manufacturers = df['manufacturer'].value_counts().head()

# Create a bar chart of the number of products for each manufacturer.
plt.bar(top_manufacturers.index, top_manufacturers.values)

# Customize the bar chart.
plt.xlabel('Manufacturer')
plt.ylabel('Number of Products')
plt.title('Most Popular Manufacturers')
plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt

no_of_reviews=df['number_of_reviews'].value_counts()
sorted_no_of_reviews=no_of_reviews.sort_values(ascending=False)
review_analysis=sorted_no_of_reviews.head()
review_analysis
```

 2) This is the analayis of the Reviews of the dataset

```python
# Create a list of pie chart slice labels.
labels = review_analysis.index.to_numpy()

# Create a list of pie chart slice values.
values = review_analysis.to_numpy()

# Create a pie chart of the number of reviews for each rating.
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart of Number of Reviews")
plt.show()
```
